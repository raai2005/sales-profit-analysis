Find which factors (discount, quantity, category, region) most affect profit.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'..\data\CleanedSalesData.csv')
```

```python
corr = df[['Sales', 'Quantity', 'Discount', 'Profit']].corr()
```

```python
plt.figure(figsize=(15,5))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Between Key Matrics")
plt.show()
```

```python
sns.scatterplot(x='Discount', y='Profit', data=df, alpha=0.6)
plt.title("Discount vs Profit Relationship")
plt.show()
```
